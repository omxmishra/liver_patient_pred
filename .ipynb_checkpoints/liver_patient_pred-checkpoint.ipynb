# Domain Analysis

**Age :** Age is the Age of the Patient in Year

**Gender :** Gender is the sex of Patient

**Total_Bilirubin :** Bilirubin is a yellow pigment produced during the breakdown of red blood cells. Elevated levels of total bilirubin in the blood can indicate liver problems.

**Direct_Bilirubin :** Direct bilirubin refers to the conjugated form of bilirubin that is excreted from the liver into the bile. Elevated levels of direct bilirubin can be an indication of liver diseases or bile duct obstruction.

**Alkaline_Phosphatase :** Alkaline phosphatase (ALP) is an enzyme . Elevated levels of ALP in the blood can indicate liver disorders.

**Alamine_Aminotransferase (ALT) :** ALT is an enzyme found primarily in the liver. Increased levels of ALT can indicate liver damage .

**Aspartate_Aminotransferase (AST) :** AST is an enzyme . Elevated levels of AST can indicate liver damage, heart problems, or muscle injury.

**Total_Proteins :** Total proteins refer to the total amount of proteins present in the blood, including albumin and globulins. Abnorml levels of total proteins can indicate various medical conditions, including liver or kidney diseases.

**Albumin :** Albumin is a protein synthesized by the liver. Low levels of albumin in the blood can be indicative of liver disease or malnutrition.

**Albumin_and_Globulin_Ratio :** The albumin and globulin ratio is calculated by dividing the albumin level by the globulin level. This ratio can provide additional information about liver function .

**Target :** The patient have Liver Disease or not.

## Missing Value Handling

Only Albumin_and_Globulin_Ratio has missing values (4 entries)

Dataset is small, dropping rows reduces data

Median is used because:

1)Robust to outliers

2)Suitable for skewed medical data

```python
df.describe()
```

```python
df['Albumin_and_Globulin_Ratio'].fillna(df['Albumin_and_Globulin_Ratio'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Target'].value_counts()
```

```python
df['Target'] = df['Target'].apply(lambda x: 1 if x == 1 else 0)
df['Target'].value_counts()
```

### Target Encoding

Original labels:

1 = Liver Disease

2 = No Liver Disease

Converted to binary:

1 = Liver 

0 = No Liver

```python
plt.figure(figsize=(6,4))

ax = sns.countplot(x='Target', data=df)
ax.set_xticklabels(['No Liver Disease', 'Liver Disease'])

total = len(df)

for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    ax.annotate(percentage,
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')

plt.title("Class Distribution of Target Variable")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df['Age'], kde=True, bins=20)

plt.title("Overall Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.kdeplot(data=df, x='Age', hue='Target', fill=True, common_norm=False, alpha=0.5)

plt.title("Age Distribution by Liver Disease Status")
plt.xlabel("Age")
plt.ylabel("Density")

plt.show()
```
